```python
x = [-1.94, -1.44, 0.93, 1.39]
y = [1.02, 0.59, -0.28, -1.04]
```

Vamos criar o sistema $Ma = w$
onde

$V$ é a matriz em que cada coluna é o valor das funções $g_i$ calculadas nos $x$

$M = V^TV$

$w = V^Ty$

```python
VT = numpy.array([g0(x),g1(x)])
```

```python
print(VT)
```

```python
V = VT.T  #determinar a matriz transposta de V
print(V)
```

```python
M = VT @ V
print(M)
```

```python
w = VT @ y
print(w)
```

```python
a = numpy.linalg.solve(M, w.T)
print(a)
```
